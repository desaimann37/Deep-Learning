```python
# Gradient Descent and cost function : 

import numpy as np
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len(x)
    learning_rate = 0.08

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])

        m_derivative = -(2 / n) * sum(x * (y - y_predicted))
        b_derivative = -(2 / n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * m_derivative
        b_curr = b_curr - learning_rate * b_derivative
        print("m {} , b {} , cost {} , iteration {}".format(m_curr, b_curr, cost , i))

x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

gradient_descent(x, y)
```

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
X
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35063668
# and let's assume the starting value for intercept b = 0
y_pred = ((78.35063668 * X) + 0).reshape(4)
```

```python
plt.scatter(X , y)
plt.plot(X , reg.predict(X) , color='red' , label='OLS')
plt.plot(X , y_pred , color='blue' , label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35063668
b = 0
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
print("Loss Value : {}".format(loss_slope))

# learning_rate = 0.125 --> Ideal Scenario
learning_rate = 0.1 

step_size = loss_slope * learning_rate
print("Step Size : {}".format(step_size))

#Calculating the new intercept : 
b = b - step_size
print("New intercept(b) Value : {}".format(b))
```

```python
y_pred1 = ((m * X) + b).reshape(4)
plt.scatter(X , y)
plt.plot(X , reg.predict(X) , color='red' , label='OLS')
plt.plot(X , y_pred1 , color='blue' , label='b = {}'.format(b))
plt.plot(X , y_pred , color='green' , label='b = 0')
plt.legend()
plt.show()
```
