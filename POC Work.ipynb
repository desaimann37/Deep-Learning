## 4.2 All Links by site wise

1.   Bloomberg
2.   LiveMint
3.   Moneycontrol



```python
def is_google_link(link):
    # Check if the link leads to a Google domain
    google_domains = ['support.google.com', 'accounts.google.com', 'www.google.com', 'policies.google.com', 'maps.google.com']
    return any(domain in link for domain in google_domains)

def search_on_web(queries):

    results_by_site = {}  # Store search results grouped by site

    for query in queries:
        # Define the search engine URLs for specific sites
        search_urls = [
            ("Bloomberg", f"https://www.google.com/search?q=site:bloomberg.com+{query.replace(' ', '+')}+financial+latest+news"),
            ("LiveMint", f"https://www.google.com/search?q=site:livemint.com+{query.replace(' ', '+')}+financial+latest+news"),
            ("Moneycontrol", f"https://www.google.com/search?q=site:moneycontrol.com+{query.replace(' ', '+')}+financial+latest+news"),
            # ("The Financial Express", f"https://www.google.com/search?q=site:financialexpress.com+{query.replace(' ', '+')}+financial+latest+news"),
            # ("CNBC", f"https://www.google.com/search?q=site:cnbc.com+{query.replace(' ', '+')}+financial+latest+news"),
            # ("Investing.com", f"https://www.google.com/search?q=site:investing.com+{query.replace(' ', '+')}+financial+latest+news"),
            # ("Investopedia", f"https://www.google.com/search?q=site:investopedia.com+{query.replace(' ', '+')}+financial+latest+news")
        ]

        # Track seen links to avoid duplicates
        seen_links = set()

        for site_name, search_url in search_urls:
            # Send an HTTP GET request to the search engine
            response = requests.get(search_url)

            # Check if the request was successful (status code 200)
            if response.status_code == 200:
                # Parse the HTML content of the search results page
                soup = BeautifulSoup(response.text, 'html.parser')

                # Find all anchor elements (<a>) containing links and their titles
                search_results = soup.find_all('a', href=True)

                # Extract links and titles
                site_results = []  # Store results for current site
                for result in search_results:
                    title = result.text.strip()
                    link = result['href']
                    if link.startswith('/url?'):
                        link = link.split('=')[1].split('&')[0]  # Extract the actual URL from the Google redirect
                    if link.startswith(('http', 'https')) and not is_google_link(link):
                        site_results.append({
                            'query': query,
                            'title': title,
                            'link': link
                        })

                # Add site results to the dictionary
                if site_name in results_by_site:
                    results_by_site[site_name].extend(site_results)
                else:
                    results_by_site[site_name] = site_results

            else:
                print("Failed to retrieve search results for:", query)

    return results_by_site

# Call the search_on_web function with the search_queries list
search_results = search_on_web(search_query)

# Print the extracted data grouped by site
for site_name, results in search_results.items():
    print(f"Search Results for '{search_query[0]}' ({site_name}):")
    for result in results:
        print(f"{result['title']}: {result['link']}")
    print()
```
